```python
# 1. Install YOLO
!pip install ultralytics
from ultralytics import YOLO

# 2. Load the base YOLOv8 model
model = YOLO('yolov8n.pt')

# 3. Train the model!
# The data.yaml file was automatically created by Roboflow.
# Let's tell YOLO to look for it inside the dataset folder.
import glob
yaml_file = glob.glob('*/data.yaml')[0]

results = model.train(
    data=yaml_file,
    epochs=50,       # Number of times it studies the images (50 is good)
    imgsz=640,       # Image size
    device=0,        # This tells it to use the GPU
    name='laparoscopy_model'
)
```

```python
# Convert the trained model into the C# ONNX format
model.export(format='onnx')
```

```python
# 1. Ask the AI to look at your image
results = model.predict(source='/content/runs/detect/laparoscopy_model/val_batch1_pred.jpg')

# 2. Get the list of bounding boxes it found
boxes = results[0].boxes

# 3. Run our logic to check for issues!
if len(boxes) == 0:
    print("✅ No issues found. The image is clean.")
else:
    print(f"⚠️ ALERT: Found {len(boxes)} issue(s)!")

    # 4. Loop through every box to see the Confidence Score
    for box in boxes:
        confidence = box.conf[0].item() * 100  # Convert decimal to percentage
        class_id = int(box.cls[0].item())      # Get the ID of the disease
        class_name = model.names[class_id]     # Get the actual name (e.g., 'Polyp')

        print(f"   -> Detected: {class_name} (Confidence: {confidence:.1f}%)")
```
